```python
def validate_recharge(amt):
    valid_plans=[199,299,399,599]
    if amt<50:
        print("recharge amount must be atleast 50 rupees")
        return False
    if amt not in valid_plans:
        print("invalid plan selected")
        print("available plans: 199,299,399,599")
        return False
    return True
while True:  #retry using while loop
    amt=int(input("enter recharge amount:"))
    if validate_recharge(amt):
        print("recharge successful!")
        break
    else:
        print("please try again.\n")
```

```python
def check_inventory(products):
    for product,stock in products.items():
        if stock<15:
            print(product,":reorder alert")
        else:
            print(product,":stock ok")
inventory={   #dictionary of produts and stock
    "rice":10,
    "wheat":25,
    "sugar":12,
    "oil":30,
    "salt":8
}
check_inventory(inventory)
```

Problem 3

```python
def process_result(marks):
    total=0
    for mark in marks:
        total+=mark  # to calculate total
    average=total/len(marks)
    if average>=50:
        res="pass"
    else:
        res="fail"
    return average,res
student_marks=[60,45,70,55]
average,status=process_result(student_marks)
print("average marks:",average)
print("result:",status)
    
```

```python
def calculate_fare(dis,peak):
    base_fare=50
    per_km_rate=12
    total_fare=base_fare+(dis*per_km_rate)
    if peak.lower()=="yes":
        total_fare+=total_fare*0.25
    return total_fare
while True:  #retry option using while loop
    dis=float(input("enter distance in km:"))
    peak=input("is it peak hour? (yes/no):")
    fare=calculate_fare(dis,peak)
    print("total fare:",fare)
    retry=input("do you want to calculate again? (yes/no):")
    if retry.lower()!="yes":
        print("Thank you for using cab fare estimator")
        break
```

```python
def check_attendance(attendance_lst):
    present_days=0
    for day in attendance_lst:  #loop to count no of present days
        if day=="P":
         present_days+=1
    total_days=len(attendance_lst)
    percentage=(present_days/total_days)*100
    if percentage>=75:
        status="eligible"
    else:
        status="not eligible"
    return percentage,status
attendance=["P","P","A","P","A","P","P","P"]
percentage,result=check_attendance(attendance)
print("attendance percentage:",percentage,"%")
print("eligibility status:",result)
```

```python
def check_password(password):  #function to check password strength
    has_digit=False
    has_special=False
    special_chars="@#$"
    for char in password:  #loop over each character in password
        if char.isdigit():
            has_digit=True
        if char in special_chars:
            has_special=True
    if len(password)>=8 and has_digit and has_special:  #to check the condition
        return "strong password"
    else:
        return "weak password"
password=input("enter your password:")  #to take input from user
result=check_password(password)
print(result)
```
