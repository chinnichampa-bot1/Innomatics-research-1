Problem 1

```python
def smart_parking(capacity,logs):
    current=0
    peak=0
    for i in logs:
        if i=="IN":
            current+=1
        elif i=="OUT":
            if current>0:
                current-=1
        if current>peak:
            peak=current
    print("currently parked vehicles:",current)
    print("peak parking usage:",peak)
    if current>capacity:  #parking status is monitored
        print("parking status:exceeded capacity")
    else:
        print("parking status:available")
capacity=50
logs=["IN","IN","IN","OUT","IN","IN","OUT"]  #taking input
smart_parking(capacity,logs)
```

Problem 2

```python
def estimate_delivery(dis,traffic,weather):
    speed=20  #taking speed as 20
    base_time=(dis/speed)*60  #calculating base_time
    if traffic=="LOW" and weather=="clear":  #conditions for traffic delay and weather delay
        traffic_delay=5
        weather_delay=0
    elif traffic=="medium" and weather=="cloudy":
        traffic_delay=15
        weather_delay=5
    elif traffic=="high" and weather=="rainy":
        traffic_delay=25
        weather_delay=6
    else:
        traffic_delay=0
        weather_delay=0
    total_time=base_time+traffic_delay+weather_delay  #total time is calculated
    print("estimated delivery time:",int(total_time),"minutes")
estimate_delivery(8,"high","rainy")
```

```python
def seat_occupy(total_seats,booked_seats):
    count=len(booked_seats)  #to count the number of seats
    occupancy=(count/total_seats)*100  #calulates occupancy percentage
    print("occupancy:",int(occupancy),"%")
    if occupancy==100:  #to show the status
        print("show status:housefull")
        print("suggestion:open additional show")
    elif occupancy>=70:
        print("show status:almost full")
        print("suggestion:monitor bookings closely")
    else:
        print("show status:seats available")
total_seats=200
booked_seats=[1]*150
seat_occupy(total_seats,booked_seats)
```

```python
def classify_load(readings):
    total=0
    for i in readings:  #to calculate total
        total+=i
    avg=total/len(readings)  #to calculate average
    print("average cpu load:",int(avg), "%")
    if avg<50:
        print("server status: normal")
    elif avg<=80:
        print("server status: warning")
    else:
        print("server status: critical")
readings=[45,60,70,85,90]
classify_load(readings)
```

```python
def resource_monitor(usage):
    overused=[]
    for i, hrs in usage.items():  #to check the resources for loop is used here
        if hrs>8:
            overused.append(i)
    if overused:  #to display overused resources
        print("overused resources:",",".join(overused))
        print("energy alert:yes")
    else:
        print("overused resources:none")
        print("energy alert:no")
usage={
    "projector":6,
    "AC":9,
    "Lights":4
}
resource_monitor(usage)
```

```python
def registrations(capacity,total_reg):
    confirmed=0
    waitlist=0
    for i in range(total_reg): #for total registrations to process
        if confirmed<capacity:
            confirmed+=1
        else:
            waitlist+=1
    print("confirmed registrations:",confirmed)
    print("waitlisted users:",waitlist)
    if confirmed==capacity:  #registration status
        print("registration status: closed")
    else:
        print("registration status: open")
registrations(100,105)
        
```
