Problem 1

```python
class SmartLight:
    def __init__(self,name):
        self.name=name
        self.status="OFF"  #default will be off state
    def on(self):
        self.status="ON"
        print(f"{self.name} is {self.status}")
    def off(self):
        self.status="OFF"
        print(f"{self.name} is {self.status}")
    def display(self):
        print(f"{self.name} is {self.status}")
light=SmartLight("Bedroom light")
action="ON"
if action=="ON":  #status based on action
    light.on()
else:
    light.off()
    
```

Problem 2

```python
class Employee:
    def __init__(self,name,emp_id,dept):
        self.name=name
        self.emp_id=emp_id
        self.dept=dept
    def id_card(self):  #details of employee
        print("Employee ID card")
        print("Name:",self.name)
        print("ID:",self.emp_id)
        print("Department:",self.dept)
emp=Employee("Rahul","EMP102","IT") #create
emp.id_card()
```

Problem 3

```python
class Contact:
    def __init__(self,name,phone):
        self.name=name
        self.phone=phone
    def contact(self):  #contact information
        print("contact saved")
        print("Name:",self.name)
        print("Phone:",self.phone)
cont=Contact("Anita", "9876543210") 
cont.contact()  #displays contact information
```

Problem 4

```python
class Product:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def price_tag(self):  #product details
        print(f"Product: {self.name}")
        print(f"Price: {self.price}")
prod=Product("Headphones", 2499)
prod.price_tag() #displays product information
```

```python
class Movie:
    def __init__(self,name,rating):
        self.name=name
        self.rating=rating
    def movie_details(self):  #movie details
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating}/5")
mov=Movie("Inception",4.8)
mov.movie_details()  #displays movie information
```
