problem 3

```python
numbers=[45,22,89,10,66]
max=numbers[0]
min=numbers[0]
for i in numbers:
    if i>max:
        max=i
    if i<min:
        min=i
print("list:",numbers)
print("max:",max)
print("min:",min)
```

problem 5

```python
attendance=["P","P","A","P","P"]
count=0
for i in attendance:
    if i=="P":
        count+=1
total_days=len(attendance)
percentage=(count/total_days)*100
print("attendance percentage:",percentage)
    
```

problem 7

```python
text="pythonp"
char_count={}
for char in text:
    if char in char_count:
        char_count[char]+=1
    else:
        char_count[char]=1
print(char_count)
```
