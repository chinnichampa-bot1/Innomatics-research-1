Problem 1

```python
def post_engagement(likes_list):
    total_likes=0
    for like in likes_list:   #loop to calculate total likes
        total_likes+=like
    print("total likes:",total_likes)
    if total_likes>=1000:
        print("post status: viral post")
    else:
        print("post status: normal engagement")
likes=[200,150,300,400]   #total no of likes
post_engagement(likes)
```

Problem 3

```python
def rainfall_checker(rainfall_data,req_level):
    total=0
    for i in rainfall_data:   #to calculate total rainfall
        total+=i
    avg=total/len(rainfall_data)  #to calculate average rainfall
    print("average rainfall:",int(avg))
    if avg>=req_level:   #to check adequancy
        print("rainfall status:adequate rainfall")
    else:
        print("rainfall status:inadequate rainfall")
seasonal_rainfall=[70,75,68,75]
rainfall_checker(seasonal_rainfall,70)
```

Problem 4

```python
def detect_accounts(usernames):
    unique_users=set(usernames)
    if len(unique_users)!=len(usernames):  #to check for existence of duplicants
        print("duplicate accounts found: yes")
    else:
        print("duplicate accounts found: no")
users=["a01","a02","a01","a21"]
detect_accounts(users)
```

```python
def filter_crops(prices):
    premium_crops=[]
    for price in prices:  #loop to filter premium prices
        if price>2000:
            premium_crops.append(price)
    print("premium crops:",premium_crops)
crop_prices=[1500,2500,1800,3200]
filter_crops(crop_prices)
    
    
```

Problem 9

```python
def check_exam(attendance_list):
    total_classes=len(attendance_list)
    attended=0
    for i in attendance_list:  #to count attended classes
        if i==1:
            attended+=1
    percentage=(attended/total_classes)*100  #calculate percentage
    print("attendance percentage:",percentage)
    if percentage>=75:  #eligibility check
        print("exam eligibility: eligible")
    else:
        print("exam eligibility: not eligible")
attendance=[1, 1, 0, 1, 1]
check_exam(attendance)
```

Problem 10

```python
def calculate_bill(units):
    bill=0
    if units<=100:  #slab calculation
        bill=units*3
    elif units<=200:
        bill=(100*3)+((units-100)*5)
    else:
        bill=(100*3)+(100*5)+((units-200)*7)
    if bill<500:  #usage status
        usage_status="low usage"
    elif bill<=1500:
        usage_status="moderate usage"
    else:
        usage_status="high usage"
    return bill,usage_status
total_bill,status=calculate_bill(250)
print("total bill:",total_bill)
print("usage status:",status)
```
